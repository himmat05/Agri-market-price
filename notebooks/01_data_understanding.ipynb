# üìä Dataset Description 
The dataset represents daily agricultural market (mandi) price records collected from multiple mandis across India. Each row corresponds to the price information of a specific crop sold in a particular mandi on a given date. The dataset includes location-related details such as State, District Name, and Market Name, which identify where the transaction took place. It also contains crop-specific information like Commodity, Variety, and Grade, which describe what crop was sold and its quality. Along with these, the dataset provides market price indicators including Minimum Price, Maximum Price, and Modal Price (in Rs./Quintal), capturing the price range and the most common trading price for that day. The Price Date column indicates the exact day on which the prices were recorded. Overall, this dataset captures the spatial (location-based), qualitative (crop and grade), and temporal (date-based) factors that influence crop prices in Indian mandis.

# üéØ Model Output Description 
The output of the model is the Modal Price (Rs./Quintal) of a crop for a specific mandi on a specific date. The modal price represents the most frequently occurring price at which the crop was traded in the mandi and is considered the most reliable indicator of the actual market value for farmers and traders. Using historical mandi data, along with location details, crop characteristics, quality grades, and time-based information, the model aims to predict the expected modal price for a future or unseen date. This predicted output can help stakeholders such as farmers, traders, and agri-platforms make informed decisions regarding when and where to sell crops to achieve better returns.
