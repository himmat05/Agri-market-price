```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(r"[()/.]", "", regex=True)
)
```

```python
df.head()
```

```python
df[["min_price_rsquintal", "max_price_rsquintal", "modal_price_rsquintal"]].dtypes
```

```python
df.shape
```

```python
df["price_date"] = pd.to_datetime(df["price_date"], errors="coerce")
df["price_date"].dtype
```

```python
df.head()
```

```python
df["modal_price_rsquintal"].isna().sum()
```

```python
df = df.dropna(subset=["modal_price_rsquintal"])
df["modal_price_rsquintal"].shape
```

```python
df.shape
```

```python
df = df.dropna(subset=["price_date"])
df["price_date"].shape
```

```python
df.shape
```

```python
df = df.rename(columns={
    "sl_no": "id",
    "district_name": "district",
    "market_name": "market",
    "min_price_rsquintal": "min_price",
    "max_price_rsquintal": "max_price",
    "modal_price_rsquintal": "modal_price",
    "price_date": "date"
})
df.head()
```

```python
df.sample(5)
```

```python
df[df["min_price"].isna() | df["max_price"].isna()]
```

```python
df[df["modal_price"].isna()]
```

```python
df.loc[
    df[["min_price", "max_price", "modal_price"]].isna().any(axis=1)
]
```

```python
df.shape
```

```python
df[
    (df["min_price"] <= 0) |
    (df["max_price"] <= 0) |
    (df["modal_price"] <= 0)].shape
```

```python
df = df[
    (df["min_price"] > 0) &
    (df["max_price"] > 0) &
    (df["modal_price"] > 0)
]
df.shape
```

```python
df = df[df["min_price"] <= df["max_price"]]
```

```python
df = df[
    (df["modal_price"] >= df["min_price"]) &
    (df["modal_price"] <= df["max_price"])
]
```

```python
df.shape
```

```python
df = df.dropna(subset=["state", "district", "market", "commodity"])
```

```python
df.shape
```

```python
df = df.drop_duplicates()
df.shape
```

```python
duplicate_rows = df[
    df.duplicated(
        subset=["market", "commodity", "date", "grade", "variety", "state", "district"],
        keep=False
    )
]
```
